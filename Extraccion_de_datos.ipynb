```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Se lee el archivo excel sin índice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
#Verificar información del DataFrame
df.info()
```

```python
#Se ve la información que aparece en cada una de las variables
df.head()
```

```python
#Fltro por objeto

#Se filtra unicamente los datos que fueron vendidos por LETICIA RAMIREZ HERNANDEZ
filtro1 = df[df['NOMBRE_VENDEDOR'] == 'LETICIA RAMIREZ HERNANDEZ']
filtro1.to_csv('Filtro1.csv')
filtro1
```

```python
#Filtro por filas

#Se filtran las filas 60, 70 y 100
filtro2 = df.iloc[[60,70,100], :]  #Filas no consecutivas
filtro2.to_csv('Filtro2.csv')
filtro2
```

```python
#Filtro por columnas

#Se filtran las líneas de la fecha de documentación, el
#nombre del vendedor y el costo total por producto
filtro3 = df.iloc[:,[1,2,11]]  #Columnas no consecutivas
filtro3.to_csv('Filtro3.csv')
filtro3
```

```python
#Se lee el archivo excel sin índice
df2 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col = 0)
```

```python
#Paso 2: Aplicamos filtro de filas vs columnas

#Se filtran la cantidad y el costo total de los productos con
#las claves F37043 y F39837
filtro4 = df2.loc[['F37043','F39837'], ['CANT','COSTO_TOTAL_CALCULADO']]
filtro4.to_csv('Filtro4.csv')
filtro4
```

```python
#Filtro por cabecera

#Se filtran los primeros 15 datos de las variables
filtro5 =df.head(15)
filtro5.to_csv('Filtro5.csv')
filtro5
```

```python
#Filtro por comparación

#Se filtran el total de productros cuya cantidad es
#mayor a 70
filtro6 = df[df['CANT'] > 70]
filtro6.to_csv('Filtro6.csv')
filtro6
```

```python
#Filtro Y

#Se filtran aqueyos productos que cuyas unidades son menores a 30
#y que sean registradas el 2022-01-05
filtro7 = df[(df['CANT']<30) & (df['FECHA_DOC'] == '2022-01-05')]
filtro7.to_csv('Filtro7.csv')
filtro7
```

```python
#Filtro O

#Se filtran aqueyos productos que cuyas unidades son menores a 30
#y todos los productos que fueron registrados el 2022-01-05
filtro8 = df[(df['CANT']<30) | (df['FECHA_DOC'] == '2022-01-05')]
filtro8.to_csv('Filtro8.csv')
filtro8
```

```python
#Filtro not

#Se filtran aqueyos productos que cuyas unidades no son menores a 30
#y que no fueron registradas el 2022-01-05
filtro9 = df[~(df['CANT']<30) & ~(df['FECHA_DOC'] == '2022-01-05')]
filtro9.to_csv('Filtro9.csv')
filtro9
```
