```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Se lee el archivo excel sin índice
df = pd.read_excel('Inmuebles.xlsx', skiprows=1)
```

```python
df.info()
```

```python
filtro1 = df.iloc[:,1:8]
filtro1
```

```python
filtro4 =  df[(df['Provincia'] == 'Girona') & (df['Superficie'] > 100)]
filtro4
```

```python
filtro5 = df[(df['Precio Venta'] > 1000000) | (df['Vendedor'] == 'Cameron') | (df['Vendedor'] == 'Joaquín')]
filtro5
```

```python
filtro6 = df[~(df['Provincia'] == 'Barcelona')]
filtro6
```

```python
filtro7 = df[~(df['Operación'] == 'Alquiler') & ~(df['Tipo'] == 'Oficina')]
filtro7
```

```python
filtro1.to_csv('1.3 Filtro1.csv')
filtro4.to_csv('1.3 Filtro4.csv')
filtro5.to_csv('1.3 Filtro5.csv')
filtro6.to_csv('1.3 Filtro6.csv')
filtro7.to_csv('1.3 Filtro7.csv')
```
